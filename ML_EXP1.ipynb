#SciPy
---
SciPy in Python is an open-source library used for solving mathematical, scientific, engineering, and technical problems. It allows users to manipulate the data and visualize the data using a wide range of high-level Python commands. SciPy is built on the Python NumPy extention.

**Sub-packages of SciPy:**

>File input/output – scipy.io

>Special Function – scipy.special

>Linear Algebra Operation – scipy.linalg

>Interpolation – scipy.interpolate

>Optimization and fit – scipy.optimize

>Statistics and random numbers – scipy.stats

>Numerical Integration – scipy.integrate

>Fast Fourier transforms – scipy.fftpack

>Signal Processing – scipy.signal

>Image manipulation – scipy.ndimage

```python
#'special' class of SciPy
from scipy.special import cbrt

#Find cubic root of 27 & 64 using cbrt() function
cb = cbrt([27, 64])
#print value of cb
print(cb)
```

```python
#permutations and combinations using SciPy special class

from scipy.special import comb,perm
#find combinations of 5, 2 values using comb(N, k)
com = comb(5, 2, exact = False, repetition=True)
print(com)

#find permutation of 5, 2 using perm (N, k) function
per = perm(5, 2, exact = True)
print(per)
```

```python
#Linear algebra using SciPy
from scipy import linalg
import numpy as np

# define square matrix
two_d_array = np.array([ [4,5], [3,2] ])
#pass value to function inv()

print(linalg.det( two_d_array )) #determinant of matrix
print(linalg.inv( two_d_array )) #inverse of matrix
```

```python
# Eigen values and Eigen vectors using SciPy

arr = np.array([[5,4],[6,3]])
#pass value into function
eg_val, eg_vect = linalg.eig(arr)
#get eigenvalues
print(eg_val)
#get eigenvectors
print(eg_vect)
```

```python
#perform single and double integration using SciPy
from scipy import integrate

# take f(x) function as f
f = lambda x : x**2
#single integration with a = 0 & b = 1  
integration = integrate.quad(f, 0 , 1)
print("Single Integration: ",integration)

#import square root function from math lib
from math import sqrt
# set  fuction f(x)
f = lambda x, y : 64 *x*y
# lower limit of second integral
p = lambda x : 0
# upper limit of first integral
q = lambda y : sqrt(1 - 2*y**2)
# perform double integration
integration = integrate.dblquad(f , 0 , 2/4,  p, q)
print("Double Integration: ",integration)
```

#Scikitlearn
---
Scikit-Learn provides clean datasets for you to use when building ML models. And when I say clean, I mean the type of clean that’s ready to be used to train a ML model. The best part? The datasets come with the Scikit-Learn package itself. You don’t need to download anything. Within just a few lines of code, you’ll be working with the data.

It features various algorithms like support vector machine, random forests, and k-neighbours, and it also supports Python numerical and scientific libraries like NumPy and SciPy.

---
To show the features offered by sklearn library, we have taken a database from its inbuilt databases and performed logistic regression again offered by the inbuilt funcitons of sklearn . This shows how powerful this library is.

```python
from sklearn.datasets import load_breast_cancer  #sklearn has inbuilt test databases as powerful as real databases
from sklearn.linear_model import LogisticRegression    #Logistic Regression
from sklearn.preprocessing import StandardScaler      #inbuilt function for preprocessing
from sklearn.model_selection import train_test_split  #inbuilt function to split test and train data
```

```python
#load the breast cancer data and few EDA
cancer = load_breast_cancer()
print(cancer.DESCR)
```

```python
print(cancer.feature_names)
```

```python
#----------------Logistic Regression
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

print('Accuracy on the training set: {:.3f}'.format(log_reg.score(X_train,y_train)))
print('Accuracy on the training set: {:.3f}'.format(log_reg.score(X_test,y_test)))

#It seems as it does better than KNN
```
